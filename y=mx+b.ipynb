```python
x = np.random.rand(50) * 10
y = -3*x + 5 + np.random.rand(50)
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1,1), y)
```

```python
x_predictions = np.linspace(0,10,50)
y_predictions = model.predict(x_predictions.reshape(-1,1))
```

```python
plt.scatter(x,y)
plt.plot(x_predictions,y_predictions)
```
